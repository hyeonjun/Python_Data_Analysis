# Pandas의 pivot_table()

* 데이터프레임의 행과 열을 변경할 때 유용





```python
# 1. 모듈 import
import pandas as pd
import numpy as np
```

```python
# 2. 외부 자료 읽기
df = pd.read_excel("/content/drive/MyDrive/Encore/데이터분석/data/02. sales-funnel.xlsx")
df.head()
```

```python
# 데이터프레임의 내부 index를 'Name' 칼럼으로 변경
pd.pivot_table(df, index=['Name'])
```

```python
# 데이터프레임의 내부 index를 'Name' 칼럼등 여러개로 변경
pd.pivot_table(df, index=['Name', 'Rep', 'Manager'])
```

```python
# 데이터프레임의 내부 index를 'Manager', 'Reg' 칼럼순으로 변경 : Manager 기준으로 그룹화 가능
pd.pivot_table(df, index=['Manager', 'Rep'])
```

```python
# 데이터프레임의 내부 index를 'Manager' 'Reg' 컬럼순으로 변경, 데이터 선택
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'])
```

```python
# 집계함수를 이용할 경우 : aggfunc = 사용할 함수
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'], aggfunc=np.sum)
```

```python
# 집계함수를 이용할 경우 : aggfunc = [사용할 함수, 사용할 함수, ..]
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'], aggfunc=[np.sum, len])
```

```python
# 컬럼명 설정
pd.pivot_table(df,
               index=['Manager', 'Rep'],
               values=['Price'],
               columns=['Product'],
               aggfunc=[np.sum, len])
```

```python
# NaN 대신 기본값 설정
pd.pivot_table(df,
               index=['Manager', 'Rep'],
               values=['Price'],
               columns=['Product'],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
# 좀 더 세분화
pd.pivot_table(df,
               index=['Manager', 'Rep', 'Product'],
               values=['Price', 'Quantity'],
               aggfunc=[np.sum, np.mean],
               fill_value=0)
```
