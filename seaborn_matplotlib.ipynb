# Seaborn과 matplotlib를 이용한 시각화

* matplotlib.pyplot : 기본적인 시각화 작업을 위한 모듈
* Seaborn : matplotlib.pyplot을 이용하여 좀더 깔끔하고 다양한 시각화를 위한 모듈


```python
# 모듈 import
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 시각화를 위한 Dummy 데이터 생성
# np.linspace(시각값, 종료값, 갯수)
x = np.linspace(0, 14, 100)
x
```

```python
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)
```

```python
# 기본 시각화모듈인 matplotlib.pyplot을 이용
plt.figure(figsize=(10, 6))
plt.plot(x, y1, y2, y3, y4)
```

```python
# Seaborn 시각화 - 배경
sns.set_style("white")
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x,y2, x,y3, x,y4)
```

```python
sns.set_style("dark")
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x,y2, x,y3, x,y4)
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x,y2, x,y3, x,y4)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x,y2, x,y3, x,y4)
sns.despine(offset=10)
```

```python
# seaborn 모듈이 제공하는 기본 데이터셋 사용
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 로그 데이터셋 시각화
sns.set_style("white")
```

```python
# boxplot - x만 설정
plt.figure(figsize=(8, 6))
sns.boxplot(x=tips['total_bill'])
```

```python
# boxplot -  x, y 설정
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# boxplot -  흡연자 (smoker : Y)
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', hue="smoker", data=tips)
```

```python
# swamplot
plt.figure(figsize=(8, 6))
sns.swarmplot(x='day', y='total_bill', data=tips, color=".5")
```

```python
# boxplot + swamplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color=".25")
```

```python
# lmplot
sns.set_style("dark")
plt.figure(figsize=(8, 6))
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
```

```python
# 흡연여부
sns.set_style("darkgrid")
plt.figure(figsize=(8, 6))
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, size=7)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8, 6))
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1", height=7)
```

```python
# heapmap
uniform_data = np.random.rand(10, 12) # 10행 12열
uniform_data
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(uniform_data)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(uniform_data, vmin=0, vmax=1)
```

```python
# flights 데이터 셋
flights = sns.load_dataset("flights")
flights.head()
```

```python
# pivot
flights = flights.pivot("month", "year", "passengers")
flights.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot=True, fmt="d")
```
